```python
import pandas as pd
import os

# Check root folder and check subfolders and existence of files

root = 'ck_fecial_data'
sub_folders = os.listdir(root)
map = {}

for sub_folder in sub_folders:
    sub_folder_path = os.path.join(root, sub_folder)
    if os.path.isdir(sub_folder_path):

        for file in os.listdir(sub_folder_path):
            # copy all the pgotos to a new folder
            file_path = os.path.join(sub_folder_path, file)
            #copy the file to a new folder
            new_file_path = os.path.join('ck_train', file)
            print(file_path, new_file_path)
            os.rename(file_path, new_file_path)
```

```python
# Create dataframe from dictionary
df = pd.DataFrame.from_dict(map, orient='index', columns=['label', 'available'])
df.index.name = 'image'
df.reset_index(inplace=True)
```

```python
df.drop(columns=['available'], inplace=True)
```

```python

# Save dataframe to csv
df.to_csv('ck_fecial_data.csv', index=False)
```

```python
import pandas as pd

# Initialize result_df with the desired columns
result_df = pd.DataFrame(columns=['Image name', 'neutral', 'happiness', 'surprise', 'sadness', 'anger', 'disgust', 'fear', 'contempt', 'unknown', 'NF'])

# Define emotion values
emotion_values = {'neutral': 10, 'happiness': 10, 'surprise': 10, 'sadness': 10, 'anger': 10, 'disgust': 10, 'fear': 10, 'contempt': 10, 'unknown': 0, 'NF': 0}

# Iterate through the entries in the original DataFrame (df) and populate result_df
for index, row in df.iterrows():
    image_name = row['image']
    label_value = str(row['label'])
    print(f'image', image_name)
    print(f'label', label_value)

    # Check if the label is in emotion_values
    if label_value in emotion_values:
        # Create a new row as a dictionary
        new_row = {'Image name': image_name, **{emotion: emotion_values[emotion] if emotion == label_value else 0 for emotion in result_df.columns[1:]}}

        # Append the new row to result_df using loc
        result_df.loc[len(result_df)] = new_row

print(result_df)
```

```python
result_df.to_csv('ck_fecial_data_refined.csv', index=False)
```
