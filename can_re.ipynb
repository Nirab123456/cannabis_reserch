```python
import requests
from io import BytesIO
from PIL import Image

# List to store information about successful downloads
successful_downloads = []

# Iterate through each row and download the image
for index, row in df.iterrows():
    name = row['name']
    img_url = row['img_url']

    # Check if the URL is valid
    if pd.notna(img_url) and isinstance(img_url, str):
        # Download the image
        response = requests.get(img_url)

        if response.status_code == 200:
            # Open the image using PIL
            image = Image.open(BytesIO(response.content))

            # Convert to 'RGB' mode if it's in 'RGBA'
            if image.mode == 'RGBA':
                image = image.convert('RGB')

            # Save the image with its corresponding name
            image.save(f"img_data/{name}.jpg")

            # Store information about successful download
            successful_downloads.append({
                'name': name,
                'img_url': img_url,
                'status': 'Success'
            })

            # Mark the row as 'Yes' in the new column
            df.at[index, 'image_available'] = 'Yes'

            print(f"Image for {name} downloaded and saved.")
        else:
            # Store information about failed download
            successful_downloads.append({
                'name': name,
                'img_url': img_url,
                'status': 'Failed'
            })

            # Mark the row as 'No' in the new column
            df.at[index, 'image_available'] = 'No'

            print(f"Failed to download image for {name}.")
    else:
        # Store information about skipped row
        successful_downloads.append({
            'name': name,
            'img_url': None,
            'status': 'Skipped'
        })

        # Mark the row as 'No' in the new column
        df.at[index, 'image_available'] = 'No'

        print(f"Skipping row {index} - Invalid or missing URL for {name}.")

print("All valid images downloaded and saved.")

# Print the recorded data
print("\nRecorded Data:")
for entry in successful_downloads:
    print(entry)
```
